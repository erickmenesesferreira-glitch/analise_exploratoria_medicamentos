#Carregando os dados

#Tratamento dos Dados

```python
# Remover duplicatas pelo nome do medicamento
print(f"Total de registros antes de remover duplicatas: {len(df):,}")

df = df.drop_duplicates(subset=['name'])

print(f"Total de registros após remover duplicatas: {len(df):,}")
print(f"Registros duplicados removidos: {248218 - len(df):,}")
```

```python
#REMOÇÃO de Colunas com Alto Índice de Nulos

#Identifica todas as colunas de efeitos colaterais
all_side_effect_cols = [col for col in df.columns if col.startswith('sideEffect')]

#Calcula a porcentagem de valores nulos para cada uma delas
null_percentage = df[all_side_effect_cols].isnull().sum() / len(df)

#Filtra as colunas para remover: aquelas com mais de 99% (0.99) de nulos
colunas_a_remover = null_percentage[null_percentage > 0.99].index.tolist()

if colunas_a_remover:
    #Remove as colunas do DataFrame original
    df = df.drop(columns=colunas_a_remover)

    print(f"Colunas removidas (com >99% de nulos): {colunas_a_remover}")
else:
    print("Nenhuma coluna de side effect encontrada com mais de 99% de valores nulos.")

print(f"Novo formato do DataFrame: {df.shape}")

```

```python
#Estruturação dos Efeitos Colaterais (Side Effects) ---


# Identifica as colunas de efeitos colaterais restantes no DF (que agora só tem 41 colunas)
side_effect_cols = [col for col in df.columns if col.startswith('sideEffect')]

# Aplica a função MELT (empilhamento)
df_effects = pd.melt(df,
                     # 'name' é o identificador (o medicamento)
                     id_vars=['name'],
                     # 'value_vars' são as colunas a serem empilhadas
                     value_vars=side_effect_cols,
                     # 'Side_Effect' será a nova coluna com os valores dos efeitos
                     value_name='Side_Effect')

# Limpeza: Remove as linhas onde a célula de Side_Effect estava originalmente vazia (NaN)
df_effects.dropna(subset=['Side_Effect'], inplace=True)
# Opcional: Garante que não há strings vazias ou apenas espaços
df_effects = df_effects[df_effects['Side_Effect'].astype(str).str.strip() != '']

print(f"\nDataFrame de Efeitos Colaterais (Limpo): {df_effects.shape}")
print("Amostra do novo DataFrame:")
print(df_effects.head())
```

```python
# --Estruturação dos Casos de Uso (Use Cases) ---
# Transforma as colunas 'use0', 'use1', etc., em um único DataFrame longo.

# Identifica as colunas de usos no DataFrame principal 'df'
use_cols = [col for col in df.columns if col.startswith('use')]

# Aplica a função MELT (empilhamento)
df_uses = pd.melt(df,
                  # 'name' é o identificador (o medicamento)
                  id_vars=['name'],
                  # 'value_vars' são as colunas de uso a serem empilhadas
                  value_vars=use_cols,
                  # 'Use_Case' será a nova coluna com os valores dos usos
                  value_name='Use_Case')

# Limpeza: Remove as linhas onde a célula de Use_Case estava vazia (NaN)
df_uses.dropna(subset=['Use_Case'], inplace=True)
# Opcional: Garante que não há strings vazias ou apenas espaços
df_uses = df_uses[df_uses['Use_Case'].astype(str).str.strip() != '']

print(f"\nDataFrame de Usos (Limpo): {df_uses.shape}")
print("Amostra do novo DataFrame:")
print(df_uses.head())
```

#Medicamentos e **Efeitos** **Colaterais**


```python
print("\n=== ANÁLISE DE EFEITOS COLATERAIS ===\n")
```

```python
# Identificar colunas de efeitos colaterais
colunas_efeitos_colaterais = [col for col in df.columns if 'sideEffect' in col]
print(f"Número de colunas de efeitos colaterais: {len(colunas_efeitos_colaterais)}")
```

```python
# Coletar todos os efeitos colaterais
todos_efeitos_colaterais = []
for coluna in colunas_efeitos_colaterais:
    efeitos_validos = df[coluna].dropna()
    todos_efeitos_colaterais.extend(efeitos_validos.tolist())
```

```python
# Contar frequência dos efeitos colaterais
from collections import Counter
contagem_efeitos_colaterais = Counter(todos_efeitos_colaterais)
```

```python
# Top 15 efeitos colaterais mais frequentes
top_efeitos_colaterais = pd.DataFrame(contagem_efeitos_colaterais.most_common(15),
                                     columns=['Efeito Colateral', 'Frequência'])
```

```python
print("Top 15 Efeitos Colaterais Mais Frequentes:")
print(top_efeitos_colaterais)
```

```python
# Visualização: Gráfico de barras dos top 10 efeitos colaterais
plt.figure(figsize=(12, 8))
top_15_efeitos = top_efeitos_colaterais.head(15)
plt.barh(top_15_efeitos['Efeito Colateral'], top_15_efeitos['Frequência'], color='lightcoral')
plt.xlabel('Frequência')
plt.ylabel('Efeito Colateral')
plt.title('Top 10 Efeitos Colaterais Mais Frequentes')
plt.gca().invert_yaxis()

plt.tight_layout()
plt.show()
```

```python
# Estatísticas dos efeitos colaterais
print(f"\nTotal de efeitos colaterais únicos: {len(contagem_efeitos_colaterais)}")
print(f"Efeito colateral mais frequente: '{top_efeitos_colaterais.iloc[0,0]}' ({top_efeitos_colaterais.iloc[0,1]:,} ocorrências)")
```

#**Efeitos** colaterais por classe **terapêutica**

```python
print("\n=== EFEITOS COLATERAIS POR CLASSE TERAPÊUTICA ===\n")
```

```python
# Selecionar as top 5 classes terapêuticas, começando com ANTI INFECTIVES
top_5_classes = therapeutic_class_dist.head(5).index.tolist()
```

```python
# Garantir que ANTI INFECTIVES seja o primeiro
if 'ANTI INFECTIVES' in top_5_classes:
    top_5_classes.remove('ANTI INFECTIVES')
    top_5_classes = ['ANTI INFECTIVES'] + top_5_classes[:4]

print("Classes terapêuticas analisadas (em ordem):")
for i, classe in enumerate(top_5_classes, 1):
    print(f"{i}. {classe}")
```

```python
# Coletar os efeitos colaterais mais comuns para cada classe
efeitos_por_classe = {}
```

```python
 # Coletar efeitos colaterais específicos desta classe
for classe in top_5_classes:
    medicamentos_classe = df[df['Therapeutic Class'] == classe]
    efeitos_classe = []
    for coluna in colunas_efeitos_colaterais:
        efeitos_validos = medicamentos_classe[coluna].dropna()
        efeitos_classe.extend(efeitos_validos.tolist())

    contagem_efeitos_classe = Counter(efeitos_classe)
    top_3_efeitos = contagem_efeitos_classe.most_common(3)
    efeitos_por_classe[classe] = top_3_efeitos
```

```python
# Visualização: Gráfico dos efeitos colaterais por classe
fig, axes = plt.subplots(2, 3, figsize=(18, 12))  # 2x3 = 6 subplots
axes = axes.flatten()

for i, (classe, efeitos) in enumerate(efeitos_por_classe.items()):
    if i < 5:  # Apenas 5 subplots
        if efeitos:  # Se há efeitos colaterais
            efeitos_df = pd.DataFrame(efeitos, columns=['Efeito', 'Frequência'])
            bars = axes[i].barh(efeitos_df['Efeito'], efeitos_df['Frequência'], color=['skyblue', 'lightcoral', 'lightgreen'])
            axes[i].set_title(f'{classe}\n({len(df[df["Therapeutic Class"] == classe])} medicamentos)', fontsize=10)
            axes[i].set_xlabel('Frequência')
            axes[i].invert_yaxis()

# Remover o último subplot
fig.delaxes(axes[5])

plt.suptitle('TOP 3 EFEITOS COLATERAIS POR CLASSE TERAPÊUTICA\n(Principais 5 Classes)',
             fontsize=18, fontweight='bold', y=0.98)
plt.tight_layout(rect=[0, 0, 1, 0.96])  # Espaço para o título
plt.show()
```

```python
#Tabela: Distribuição de Medicamentos que Induzem Hábito (Habit Forming)
print("\nTabela: Distribuição de Medicamentos que Induzem dependência (Habit Forming)")
print("-" * 70)
print(df['Habit Forming'].value_counts().to_string(header=['Contagem']))
print("-" * 70)
```

```python
# Relação entre Classe Terapêutica e Formação de Hábito ---

#Identificar as 10 Classes Terapêuticas mais comuns (para simplificar a visualização)
top_10_classes = df['Therapeutic Class'].value_counts().head(10).index.tolist()

#Filtrar o DataFrame apenas para essas classes
df_filtered = df[df['Therapeutic Class'].isin(top_10_classes)]

#Criar a Tabela Cruzada (Crosstab)
# Normalizamos para ver a porcentagem de "Yes" (Sim) em cada classe.
habit_crosstab = pd.crosstab(df_filtered['Therapeutic Class'],
                            df_filtered['Habit Forming'],
                            normalize='index') * 100

#Exibir a Tabela de Relação (Porcentagem de Sim/Não por Classe)
print("Tabela 3: Porcentagem de Medicamentos Indutores de Hábito (Yes) por Classe Terapêutica")
print("-" * 100)
print(habit_crosstab.sort_values(by='Yes', ascending=False).to_string(float_format="%.2f%%"))
print("-" * 100)

#Visualização da Relação (Gráfico de Barras Empilhadas)
#Reordenar os dados para que o 'Yes' fique mais evidente no topo
plot_data = habit_crosstab.sort_values(by='Yes', ascending=False)

# Criar o gráfico
plt.figure(figsize=(12, 7))
plot_data.plot(kind='bar', stacked=True, color=['#1f77b4', '#d62728'], ax=plt.gca())

plt.title('Percentual de Medicamentos que induzem a dependência (Sim/Não) por Classe Terapêutica', fontsize=16)
plt.xlabel('Classe Terapêutica', fontsize=12)
plt.ylabel('Percentual', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Indutor de Hábito', labels=['Não', 'Sim'], loc='lower right')
plt.tight_layout()
plt.show()
```

#Qualidade e Tratamento Final dos Dados

```python
# Análise Visual do Padrão de Nulos (Missing Data) ---
import missingno as msno
import matplotlib.pyplot as plt
# Calcula a porcentagem de nulos por coluna
null_counts = df.isnull().sum()
null_percentages = (null_counts / len(df)) * 100

# Filtra colunas com mais de 0% de nulos para análise
missing_data = null_percentages[null_percentages > 0].sort_values(ascending=False)

print("Tabela: Porcentagem de Valores Nulos por Coluna (Apenas Nulos)")
print("-" * 55)
print(missing_data.to_string(float_format="%.2f%%"))
print("-" * 55)

# Visualização: Mapa de Nulos (Matrix Plot)
# Mostra a densidade e o padrão de nulos
plt.figure(figsize=(15, 8))
msno.matrix(df, figsize=(15, 8), color=(0.4, 0.2, 0.6)) # Exibe apenas o df principal
plt.title('Padrão de Dados Faltantes no DataFrame Principal (df)', fontsize=18)
plt.show()
```

```python
#Imputação Categórica e Tratamento Final

# 9.1 Imputação das Classes Químicas e de Ação
colunas_para_imputar = ['Chemical Class', 'Action Class']

for col in colunas_para_imputar:

    df[col] = df[col].fillna('DESCONHECIDA')

print("Imputação Concluída nas Classes Química e de Ação.")
```
